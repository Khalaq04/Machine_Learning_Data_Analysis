```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt

#read the training set into a data frame
df = pd.read_csv('/home/khalaq04/Downloads/weatherAUS.csv')

#first 5 observations
df.head()
```

```python
#last 5 observations
df.tail()
```

```python
#removing useless features from the dataset
df = df.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm'],axis=1)

#sample of the new dataframe
df.head(1)
```

```python
#replacing the null values with mean values of every column
df.fillna(value=df.mean(), inplace=True)

#displaying the first five observations of the new dataframe
df.head()
```

```python
#to change the strings in the dataset to integers(i.e. to change yes to 1 and no to 0) and to replace the null values
df.RainToday = [1 if data == 'Yes' else 0 for data in df.RainToday]
df.RainTomorrow = [1 if data == 'Yes' else 0 for data in df.RainTomorrow]
df.fillna(value=df.mean(), inplace=True)

#displaying a sample of the new dataframe
df.head(1)
```

```python
#determining the number of features
no_features = len(df.iloc[:,:17].columns)
no_features
```

```python
#normalisation of the features
df.iloc[:,:] = (df - df.min())/(df.max()-df.min())

#sample of the new dataframe
df.head(1)
```

```python
#creating x,y,theta
x = df.iloc[:,1:]
ones = np.ones([len(df),1])
x = np.concatenate((ones,x),axis=1)
y = df.iloc[:,0:1].values
xt = x.T
yt = y.T
theta = np.zeros([1,no_features+1])

#storing the target
target = y

#display the size of the matrices
x.shape,y.shape,theta.shape,xt.shape,yt.shape
```

```python
#computing the sigmoid function
def sigmoid(z):
    g = 1/(1 + np.exp(-z))
    return g

#set number of epochs and the learning rate
epochs=3000
alpha=1.9

#generate cost function
def cost_function(xt,yt,theta):
    h = sigmoid(theta@xt)
    J = -(np.sum(((1-yt)*np.log(1-h)+yt*np.log(yt)))/(len(x)))
    return J

#generate gradient descent
def gradient_descent(x,y,xt,yt,theta,epochs,alpha):
    cost = np.zeros(epochs)
    for i in range(epochs):
        h = sigmoid(theta@xt)
        theta -= (alpha/len(x))*np.sum(x*(h.T-y),axis=0)
        cost[i] = cost_function(xt,yt,theta)
    return theta,cost

#use gradient descent and display the final values of theta
ftheta,cost = gradient_descent(x,y,xt,yt,theta,epochs,alpha)
ftheta.round(4)
```

```python
#predictions for the first five test values
predictions = sigmoid(ftheta@xt)
for i in range(0,5):
    print(predictions[0][i].round(5))
```

```python
#to see if it will rain tomorrow(1) or not(0) for the first five test values
final_predictions = predictions
for i in range(len(predictions[0])):
    if predictions[0,i].round(5)>=0.5:
        final_predictions[0,i]=1
    else:
        final_predictions[0,i]=0
for i in range(0,5):
    print(final_predictions[0][i])
```

```python
#to measure the accuracy(%) for the test epochs
print('Accuracy = ' + str((100 - np.mean(np.abs(final_predictions - yt))*100).round(2)) + '%')
```
