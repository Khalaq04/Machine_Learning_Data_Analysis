```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

#read the training set into a data frame
df = pd.read_csv('/home/khalaq04/Downloads/weatherAUS.csv')

#first 5 observations
df.head()
```

```python
#last 5 observations
df.tail()
```

```python
#removing useless features from the dataset
df = df.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm'],axis=1)

#sample of the new dataframe
df.head(1)
```

```python
#replacing the null values with mean values of every column
df.fillna(value=df.mean(), inplace=True)

#displaying the first five observations of the new dataframe
df.head()
```

```python
#to change the strings in the dataset to integers(i.e. to change yes to 1 and no to 0) and to replace the null values
df.RainToday = [1 if data == 'Yes' else 0 for data in df.RainToday]
df.RainTomorrow = [1 if data == 'Yes' else 0 for data in df.RainTomorrow]
df.fillna(value=df.mean(), inplace=True)

#displaying a sample of the new dataframe
df.head(1)
```

```python
#separating features and target into x and y respectively
y = df['RainTomorrow']
x = df.drop('RainTomorrow',axis=1)

#storing the target values
target = y

#implementing decision tree for classification
dtc = DecisionTreeClassifier()
dtc.fit(x,y)

#to predict the target values for the training set
predictions = dtc.predict(x)

#comparison between the actual target values and the predicted target values
compare = pd.DataFrame({'Actual Values':target,'Predicted Values':predictions})
compare.head()
```

```python
#displaying predictions for random observations
compare.sample(5)
```

```python
#to measure the accuracy(%) for the predictions
print('Accuracy = ' + str((100 - np.mean(np.abs(predictions - target))*100).round(2)) + '%')
```
